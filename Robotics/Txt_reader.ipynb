```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

def predictor(fn):
    df= pd.read_csv(fn ,delimiter=" ", skiprows=[0], header=None,)
    df = df.rename(columns={1: 'Gender', 2: 'Age', 3: 'Annual Income (k$)' ,4: 'Spending Score (1-100)'})
    df=df.drop(columns=[0])
    df=df.replace('Male',1)
    df=df.replace('Female',2)

    minobs = min(df.value_counts('Gender').values)
    df = df.groupby('Gender').sample(n=minobs, random_state=1).sample(frac=1, random_state=1)
    data= np.asarray(df)
    display(df)

    X= data[20:, 1:]
    Xmax = X.max(0)
    X = X / Xmax
    Y= data[20:, 0]

    
    #print(X)
    print(X.shape, Y.shape)

    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
    print('training size:', X_train.shape[0], 
           'testing size:', X_test.shape[0],
           'label counts:', np.unique(y_train, return_counts=True)[1])

    clf = MLPClassifier(hidden_layer_sizes=(100,20),
                        max_iter=50).fit(X_train, y_train)

    print('score train:', clf.score(X_train, y_train))
    print('score test: ', clf.score(X_test, y_test))


    with open("Log.txt", "w") as f:
        print('score train:', clf.score(X_train, y_train), file=f)
        print('score test: ', clf.score(X_test, y_test), file=f)
        print('pred prob CLF I: ', clf.predict_proba(X[:1]), file=f)
        print('pred class CLF I:', clf.predict(X[:1]), file =f)
        
    print('pred prob CLF I: ', clf.predict_proba(X[:1]))
    print('pred class CLF I:', clf.predict(X[:1]))
    
    import pickle
    pickle.dump((clf, Xmax), open('ccd.pkl', 'wb'))
    
    clf2, Xmax2 = pickle.load(open('ccd.pkl', 'rb'))
    

    print('pred class CLF II:', clf2.predict(X[:1]))
    #print(df.iloc[20:, 1:])
    print('pred prob CLF I: ', clf2.predict_proba(df.iloc[20:21, 1:]/Xmax2))

predictor("Mall.txt")
```
