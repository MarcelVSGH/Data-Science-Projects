```python
import numpy as np
import pandas as pd
df= pd.read_csv("Mall_Customers.csv")

#Preprocessing and Data Cleaning
df=df.replace('Male',1)
df=df.replace('Female',2)
df = df.rename(columns={'Genre': 'Gender'})
df=df.drop(columns=['CustomerID'])

#isplay(df)
df.to_csv('new_df.csv', index=False)

minobs = min(df.value_counts('Gender').values)
df = df.groupby('Gender').sample(n=minobs, random_state=1).sample(frac=1, random_state=1)
print(df)
data= np.asarray(df)

X= data[:, 1:]
Xmax = X.max(0)
X = X / Xmax
Y= data[:, 0]

print(Y)
print(X)
print(X.shape, Y.shape)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

print(X.shape, Y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print('training size:', X_train.shape[0], 
      'testing size:', X_test.shape[0],
      'label counts:', np.unique(y_train, return_counts=True)[1])

clf = MLPClassifier(hidden_layer_sizes=(100,20),
                    max_iter=50).fit(X_train, y_train)

print('score train:', clf.score(X_train, y_train))
print('score test: ', clf.score(X_test, y_test))
```

```python
print('pred prob: ', clf.predict_proba(X[:1]))
print('pred class:', clf.predict(X[:]))
```

```python
import pickle

pickle.dump((clf, Xmax), open('ccd.pkl', 'wb'))
```
